```python
import numpy as nm
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
data = pd.read_csv('../Data/poly-data.csv')
x = data.iloc[:,1:2]
y = data.iloc[:,2]
```

```python
lin_regs= LinearRegression()
lin_regs.fit(x,y)
```

```python
poly_regs = PolynomialFeatures(degree=3)
x_p = poly_regs.fit_transform(x)
lin_reg2 = LinearRegression()
lin_reg2.fit(x_p,y)
```

```python
plt.scatter(x,y,color="blue")
plt.plot(x,lin_regs.predict(x), color="red")
plt.title("Bluff detection model(Linear Regression)")
plt.xlabel("Position Levels")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x,y,color="blue")
plt.plot(x,lin_reg2.predict(x_p), color="red")
plt.title("Bluff detection model(Linear Regression)")
plt.xlabel("Position Levels")
plt.ylabel("Salary")
plt.show()
```
